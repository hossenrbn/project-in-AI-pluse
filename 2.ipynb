```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```python
import seaborn as sns
```

```python
data=pd.read_csv('finail_2.csv')
```

```python
data['تاریخ']=pd.to_datetime(data['تاریخ'])
```

```python
data.set_index('تاریخ',inplace=True)
```

```python
data=data[data.index.month!=3]
```

```python
def trim(x):
    return stats.trim_mean(x,.2)
```

```python
trim_mein=data.resample('W').apply(trim)
mein=data.resample('W').mean()
```

```python
plt.figure(figsize=(12,3))
plt.plot(trim_mein['قیمت کل'])
#plt.plot(data['قیمت کل'].,color='red')

plt.plot(mein['قیمت کل'],color='green')

plt.show()
```

-------------------       نمایش داده ها به صورت هفتگی      ------------------ 

```python
for i in range(1,23):

    a=data[data['zone']==i]
    print('-------------------------',i,'منطقه','-------------------')
    print('کمترین قیمت واحد ملیون',a['قیمت واحد'].min()/1000000,'------','بیشترین قیمت واحد ملیون',a['قیمت واحد'].max()/1000000)
    print('     ','میانگین trim قیمت واحد به ملیون ',trim(a['قیمت واحد'])/1000000)
    print('کمترین قیمت کل میلیارد',a['قیمت کل'].min()/1000000000,'------','بیشترین قیمت کل میلیارد',a['قیمت کل'].max()/1000000000)

    print('     ','میانگین trim قیمت کل به میلیارد ',trim(a['قیمت کل'])/1000000000)

    trim_mean=a.resample('W').apply(trim)
    mein=a.resample('W').mean()
    plt.figure(figsize=(12,3))
    plt.plot(trim_mean['قیمت کل'])
    plt.plot(mein['قیمت کل'],color='black')

    plt.show()
```
