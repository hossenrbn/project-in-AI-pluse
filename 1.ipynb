-------------------       نمایش داده ها به صورت هفتگی      ------------------ 

```python
for i in range(1,23):

    a=data[data['zone']==i]
    print('-------------------------',i,'منطقه','-------------------')
    print('کمترین قیمت واحد ملیون',a['قیمت واحد'].min()/1000000,'------','بیشترین قیمت واحد ملیون',a['قیمت واحد'].max()/1000000)
    print('     ','میانگین هتدسی قیمت واحد به ملیون ',gmean(a['قیمت واحد'])/1000000)
    print('کمترین قیمت کل میلیارد',a['قیمت کل'].min()/1000000000,'------','بیشترین قیمت کل میلیارد',a['قیمت کل'].max()/1000000000)

    print('     ','میانگین هتدسی قیمت کل به میلیارد ',gmean(a['قیمت کل'])/1000000000)

    g_mein=a.resample('W').apply(gm)
    mein=a.resample('W').mean()
    plt.figure(figsize=(12,3))
    plt.plot(g_mein['قیمت کل'])
    plt.plot(mein['قیمت کل'],color='black')

    plt.show()
```
